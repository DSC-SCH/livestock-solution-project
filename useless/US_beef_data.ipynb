## 미국소고기 전체 지역(All state) 생산량 데이터 전처리

```python
df_production.head()
```

```python
df_production.columns
```

```python
# 필요한 칼럼들만 추출 
used_columns = ['Year','Period','State','Commodity','Value($)']
df_production = df_production[used_columns].copy()
df_production.head()
```

```python
# 데이터 칼럼명 수정
df_production.columns = ['연도','월','국가','상품','가격($달러)']
df_production.head()
```

```python
# 칼럼별 데이터 type 보기
df_production.info()
```

```python
# 월 칼럼 숫자로 바꿔주기 위해서 '월' 칼럼 값 종류 보기
df_production['월'].unique()

# YEAR 은 뭐지..?
```

```python
df_production.loc[df_production['월'] == 'YEAR']

# 연도별 집계 자료였다. 필요없으니 YEAR값을 갖고있는 행 제외한 데이터프레임 따로 저장
```

```python
df_production = df_production.loc[df_production['월'] != 'YEAR'].copy()

# 잘 제거됬는지 확인
df_production.loc[df_production['월'] == 'YEAR']
```

```python
# 영어로 된 월(Month)에 해당하는 숫자로 바꿔주기
df_production.loc[df_production['월'] == 'JAN', '월'] = 1
df_production.loc[df_production['월'] == 'FEB', '월'] = 2
df_production.loc[df_production['월'] == 'MAR', '월'] = 3
df_production.loc[df_production['월'] == 'APR', '월'] = 4
df_production.loc[df_production['월'] == 'MAY', '월'] = 5
df_production.loc[df_production['월'] == 'JUN', '월'] = 6
df_production.loc[df_production['월'] == 'JUL', '월'] = 7
df_production.loc[df_production['월'] == 'AUG', '월'] = 8
df_production.loc[df_production['월'] == 'SEP', '월'] = 9
df_production.loc[df_production['월'] == 'OCT', '월'] = 10
df_production.loc[df_production['월'] == 'NOV', '월'] = 11
df_production.loc[df_production['월'] == 'DEC', '월'] = 12
```

```python
# 변환 완료
df_production['월'].unique()
```

```python
df_production.info()
```

```python
# 월 데이터 유형을 int형으로 바꿔주자
df_production['월'] = df_production['월'].astype(int)
```

```python
# 가격칼럼도 ,을 제거하고 int형으로 바꿔주자
df_production['가격($달러)'] = df_production['가격($달러)'].str.replace(",","")
df_production['가격($달러)'] = df_production['가격($달러)'].astype(float)
```

```python
df_production.head()
```

```python
# 가격 단위를 만 달러($)로 변환
df_production['가격($달러)'] = df_production['가격($달러)'].div(10000)
df_production.head()
```

```python
df_production = df_production.rename(columns={'가격($달러)':'가격($만달러)'})
```

```python
df_production.head()
```

```python
# 2003 ~ 20219년도 데이터만 추출
df_production = df_production.loc[(df_production['연도'] < 2020) & (df_production['연도'] >2002)].copy()

# 연도와 월별로 재정렬
df_production = df_production.sort_values(by=['연도','월'])
```

```python
df_production.head()
```

```python
# 인덱스 reset
df_production = df_production.reset_index(drop=True).copy()
df_production.head()
```

```python
df_production.to_csv("us_beef_production.csv")
```
