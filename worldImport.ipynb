갑자기 2015년에 소고기 수입량이 증가한 이유가 무엇일까?   
냉장육의 수입도 증가했고 냉동육의 수입도 증가했다  
1) 한우와 국산 돼지고기의 가격상승  
한우 등심(100g)은 1만1900원 -> 호주산 냉장 등심은 5000원  
같은 무게 한우 안심은 1만2800원인데, 호주산 안심은 6000원   
   
2) 메르스의 여파  
2015년 6월경 메르스로 인해 소비가 위축되자, 대형마트는 소비 촉진을 위해 대대적인 수입육 행사를 진행.   
그 덕분에 수입육의 양도 증가함.    
   
출처)https://biz.chosun.com/site/data/html_dir/2015/08/25/2015082501021.html

### (4). 누적영역형그래프 

```python
colors = ["#AA2848", "#26852e", "#2848AA"]
plt.plot([],[],color="#AA2848", label =worldImport2013['total'], linewidth=3) #빨간색 
plt.plot([],[],color="#26852e", label = worldImport2013['frozen'], linewidth=3) #초록색 
plt.plot([],[],color="#2848AA", label =worldImport2013['refrigeration'], linewidth=3) #파란색 

plt.xlabel ('month')
plt.ylabel ('import')
plt.title("2013 frozen and refrigeration meat status")

plt.stackplot(worldImport2013['month'], worldImport2013['total'],worldImport2013['frozen'],worldImport2013['refrigeration'],colors=colors)
plt.xlim([1,12])
#plt.legend()
plt.show()
```

```python
colors = ["#AA2848", "k", "#2848AA"] #k는 검정색 
plt.plot([],[],color="#AA2848", label =worldImport2014['total'], linewidth=3)
plt.plot([],[],color="k", label = worldImport2014['frozen'], linewidth=3)
plt.plot([],[],color="#2848AA", label =worldImport2014['refrigeration'], linewidth=3)

plt.xlabel ('month')
plt.ylabel ('import')
plt.title("2014 frozen and refrigeration meat status")

plt.stackplot(worldImport2014['month'], worldImport2014['total'],worldImport2014['frozen'],worldImport2014['refrigeration'],colors=colors)
#원래는 frozen 이랑 refrigeration 이랑 더해서 total 값이 나오는데 그래프 상에서 보면 total(파란색)이 조금 더 나와있음 
plt.xlim([1,12])
#plt.legend() : 위에 설정한 label이 데이터프레임의 열이라서 범례를 넣으면 그 값들도 같이 출력됨 
plt.show()
```
