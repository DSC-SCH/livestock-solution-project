---
* 필요한 칼럼은 날짜(시간), 기온, 습도와 관련된 칼럼이다.
    * 날짜(시간) 관련 칼럼
        - 일시
    * 기온 관련 칼럼
        - 평균기온, 최저기온, 최고기온
        - 합계 일조 시간, 1시간 최다일사량, 합계 일사
        - 평균 지면온도
        - *`일사(日射): 태양의 복사 에너지가 땅에 닿았을 때의 세기`*
    * 습도 관련 칼럼
        - 일강수량, 1시간 최다강수량, 10분 최다 강수량
        - 최소 상대습도, 평균 상대습도, 
    * 기타 필요할 수도 있을 칼럼
        - 도, 지역

```python
# 필요한 칼럼 추출
columns = ['도','지역','일시','평균기온(°C)', '최저기온(°C)','최고기온(°C)','합계 일조 시간(hr)',
          '1시간 최다일사량(MJ/m2)','합계 일사(MJ/m2)','평균 지면온도(°C)','일강수량(mm)',
          '1시간 최다강수량(mm)','10분 최다 강수량(mm)','최소 상대습도(%)','평균 상대습도(%)']
weather = weather[columns].copy()
weather.head()
```

```python
# 일시 데이터가 객체 이므로 Datetime으로 변환
weather['일시'] = pd.to_datetime(weather['일시'])
weather.dtypes
```

```python
# 일시 기준으로 2013-01-01 ~ 2018-12-31 데이터만 추출
df_weather = weather[weather['일시'].isin(pd.date_range(start='2013-01-01', end='2018-12-31'))]

# 인덱스 리셋팅
df_weather = df_weather.reset_index(drop=True)
df_weather.head()
```

```python
# 추출한 데이터 결측치 다시 집계해보기
df_weather.isnull().sum()
```

- 똑같은 현상을 의미하는 여러가지 칼럼 중 결측치가 많은 칼럼 삭제
    * 기온을 나타내는 칼럼 중 **1시간 최다일사량, 합계 일사** 삭제
    * 우리가 진짜 원하는 칼럼은 **`습도`** 이므로 비온날**만** 기록되는 강수량 관련 칼럼 삭제
        * **일강수량, 1시간 최대강수량, 10분 최다 강수량** 삭제

```python
df_weather = df_weather.drop(['1시간 최다일사량(MJ/m2)','합계 일사(MJ/m2)','일강수량(mm)','10분 최다 강수량(mm)',
                '1시간 최다강수량(mm)'], axis=1).copy()
df_weather
```

```python
# 다시 결측치 확인
df_weather.isnull().sum()
```

```python
# 칼럼명에 측정단위 문자가 들어가 있어 코드 작성시 불편함 우려로 인해 칼럼 이름 재작성
df_weather = df_weather.rename(columns={'평균기온(°C)':'평균기온','최저기온(°C)':'최저기온','최고기온(°C)':'최고기온',
                          '합계 일조 시간(hr)':'합계일조시간','평균 지면온도(°C)':'평균지면온도',
                          '최소 상대습도(%)':'최소상대습도(%)','평균 상대습도(%)':'평균상대습도(%)'}).copy()
df_weather.head()
```

```python
# '연도', '월' 2개의 새로운 칼럼 만들어주기
df_weather['연도'] = df_weather['일시'].dt.year
df_weather['월'] = df_weather['일시'].dt.month
df_weather.head()
```

```python
# 연도, 월 별로 그룹핑하여 연도, 월별 기온~습도 평균값 집계
weather_g = df_weather.groupby(['연도','월'])['평균기온','최저기온','최고기온','합계일조시간','평균지면온도','최소상대습도(%)',
                              '평균상대습도(%)'].mean().round()
weather_g = weather_g.reset_index()
weather_g
```

## 한우 경매 가격 데이터 전처리

- 연도,월 별로 그룹핑한 평균값 데이터(변수명: weather_g)에 맞게 한우 경매 가격 데이터 포맷 수정

```python
cow_price.shape
```

```python
cow_price.columns
```

```python
# 필요한 칼럼만 추출
columns = ['경매시간','경매 구분 명','도매시장명','한우 평균경매 금액','돼지 탕박 경매 금액']
cow_price = cow_price[columns].copy()
```

```python
cow_price.head()
```

```python
# "경매용+지육상장 and 도매시장명이 전국 and 한우 평균경매 금액이 0이 아닌" 데이터들만 추출
cow_price = cow_price[(cow_price['경매 구분 명'] == '경매용+지육상장') & (cow_price['도매시장명'] == '전국')].copy()
df_price = cow_price[cow_price['한우 평균경매 금액'] != 0]
df_price
```

```python
df_price.dtypes
```

```python
# 경매시간 dtype을 str로 바꾸고 Datetime으로 바꾸기
df_price['경매시간'] = df_price['경매시간'].astype(str)
df_price['경매시간'] = pd.to_datetime(df_price['경매시간'])
df_price.dtypes
```

```python
# '연도', '월' 두 개의 새로운 칼럼 만들기
df_price['연도'] = df_price['경매시간'].dt.year
df_price['월'] = df_price['경매시간'].dt.month
df_price.head()
```

```python
# 경매시간별로 정렬
df_cow_price = df_price.sort_values(by=['경매시간']).reset_index(drop=True)
df_cow_price
```

```python
# 연도, 월 그룹핑해서 한우 평균경매 금액, 돼지탕박 경매금액의 합, 평균 데이터프레임 2개 생성
price_g_sum = pd.DataFrame(df_cow_price.groupby(['연도','월'])['한우 평균경매 금액','돼지 탕박 경매 금액'].sum()).reset_index()
price_g_mean = pd.DataFrame(df_cow_price.groupby(['연도','월'])['한우 평균경매 금액','돼지 탕박 경매 금액'].mean()).reset_index()
price_g_sum.head()
```

```python
# 금액 소수점 없애기
price_g_mean = price_g_mean.round()
price_g_mean.head()
```

## 연도, 월 별 날씨 데이터와 가격 데이터 상관성 비교

- 연도,월별 날씨 데이터의 여러가지 기온,습도 요소와 비교
    * 연도, 월별 한우, 돼지 탕박 각 합계 금액과의 비교
    * 연도, 월별 한우, 돼지 탕박 각 평균 금액과의 비교

```python
# 날씨 데이터 & 한우,돼지 합계 경매금액과의 비교
merge_df1 = pd.merge(weather_g, price_g_sum, how='outer', on=['연도','월'])
corr_data = merge_df1.corr()
corr_data.style.background_gradient()
```

```python
# 날씨 데이터 & 한우,돼지 평균 경매금액과의 비교
merge_df2 = pd.merge(weather_g, price_g_mean, how='outer', on=['연도','월'])
corr_data = merge_df2.corr()
corr_data.style.background_gradient()
```

---
- 두 개의 상관관계 그래프로 이끌어 낼 수 있는 결론
    1. 한우, 돼지탕박 경매가격 **합계** 그래프에서는 그렇게 큰 상관성을 이끌어내지 못했다.
        - 하나 이끌어낸다면 **한우 경매금액** 과 **돼지 탕박 경매금액**은 0.9라는 매우 강한 양의상관관계를 보였다. => 역시 돼지, 소 둘 중 하나가 상승한다면 나머지도 상승하는 듯!
    2. 하지만 한우, 돼지탕박 경매가격 **평균** 그래프에서는 꽤 상관성이 보였다.
        - 돼지 탕박 경매 금액이 다른 기온,습도와의 상관성을 바라보자.
            - ``기온, 지면온도가 높을 수록, 상대습도가 높을수록 돼지 탕박 경매 금액이 올라가는 양의상관관계를 보였다.``
        - ``한우 평균경매 금액은 돼지 탕박에 비해 기온과 습도의 영향을 덜 받기는 한다. 하지만 상대습도와 양의상관관계를 약간 보임``
        - 합계 그래프 때보다는 약하지만 그래도 한우 평균경매 금액은 돼지탕박 경매금액과 0.45라는 양의상관관계를 보인다.
        
    3. 한가지 의문 점 : 합계로 집계할 때는 **'연도' 와 '소, 돼지 경매 가격'** 이 완전 음의 상관관계이다가 평균 집계로 하면 완전 반대의 양의 상관관계를 띰..개인적인 생각으로는 평균으로 집계하는 것이 맞을 것 같습니다... 연도가 지날 수록 소의 경매금액은 당연히 올라가지 않을까요..? 저의 개인 상식적인 사고에 의해서 그렇게 판단이 되는데 이에 대해 의견 부탁드리겠습니다..!
