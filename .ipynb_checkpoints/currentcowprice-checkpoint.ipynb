```python

auctionprice=pd.read_csv("C:/Users/jimin/Desktop/EDA/01. 축산물 실시간 경매시황중계정보.csv",engine="python")
```

```python
cowauction = auctionprice[['경매시간','경매 구분 명','도매시장명','한우 평균경매 금액']]
```

```python
cowauction.head(5)
```

```python
cowauction=cowauction.rename({'경매시간':'date','경매 구분 명':'auctionSection','도매시장명':'marketName','한우 평균경매 금액':'averageCost'},axis='columns')
cowauction.tail(5)
```

```python
cowauction.info()
```

```python
subset1= cowauction['marketName']=="전국"
cowauction= cowauction[subset1]
cowauction
```

```python
subset2=cowauction['auctionSection']=='경매용+지육상장' 
cowauction = cowauction[subset2] 
cowauction
cowauction['date'] = cowauction['date'].astype(str)
cowauction.head(5)


```

```python
cowauction.info()
```

```python
yearmonthdate=pd.DataFrame(cowauction.loc[:,'date'])
yearmonthdate
```

```python
yearmonthdate=yearmonthdate.reset_index()
del yearmonthdate['index']
yearmonthdate
```

```python
year = pd.DataFrame(columns=['year'])
for i in range(0,1030):
    yeardata =yearmonthdate['date'][i][0:4]
    year=year.append(pd.DataFrame([yeardata],columns = ['year']),ignore_index=True)
year
```

```python
month = pd.DataFrame(columns=['month'])
for i in range(0,1030):
    monthdata =yearmonthdate['date'][i][4:6]

    month=month.append(pd.DataFrame([monthdata],columns = ['month']),ignore_index=True)
month
```

```python
date = pd.DataFrame(columns=['date'])
for i in range(0,1030):
    datedata =yearmonthdate['date'][i][6:8]

    date=date.append(pd.DataFrame([datedata],columns = ['date']),ignore_index=True)
date
```

```python
yearmonthdate=pd.concat([year,month],join="inner", axis=1)
```

```python
yearmonthdate
```

```python
yearmonthdate=pd.concat([yearmonthdate,date],join="inner", axis=1)
yearmonthdate
```

```python

yearmonthdate
```

```python
averageCost=pd.DataFrame(cowauction['averageCost'])
averageCost
```

```python
current=pd.concat([yearmonthdate,averageCost],join="outer", axis=1)
current
```

```python
grouped = current["averageCost"].groupby([current["year"],current["month"]])
yearmonth_grouped= grouped.mean()
yearmonth_grouped
```

```python
pd.merge(yearmonthdate,averageCost,how="inner")
```
