```python
#iqr로 이상치 확인해보기
```

```python
def outliers_iqr(column):
        quartile_1, quartile_3 = np.percentile(column,[25,75])
        iqr = quartile_3 - quartile_1
        lower_bound = quartile_1 - (iqr * 1.5)
        upper_bound = quartile_3 + (iqr * 1.5)
        
        return np.where((column > upper_bound) | (column < lower_bound),False,True)
        #column 의 값이 upper_bound 보다 크거나 lower_bound보다 작으면 False출력 
        
    
```

```python
outliers_iqr(usbeef['price'])
```

```python
usbeef.plot.scatter(x='yearmonth',
                   y='price',
                   s=50,
                   c='blue',
                   alpha=0.5)
plt.show()
```
